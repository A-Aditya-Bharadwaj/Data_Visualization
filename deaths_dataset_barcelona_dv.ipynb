
## Working with barcelona dataset

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
plt.style.use("seaborn")
```

```python
from matplotlib import style
print(plt.style.available)
```

### Importing deaths.csv file

```python
df = pd.read_csv("datasets/Barcelona Datasets/deaths.csv")
df
```

### Creating a dataframe with District Name as index and the total number of deaths happend in Barcelona as the columns 

### Ploting two box plot
##### - In the first box plot  we analyse the total number of deaths happend in the country
##### - In the secound box plot we analyse the total number of deaths happend in each district

```python
df_deaths = df.groupby("District.Name")["Number"].nunique()
df_deaths
```

```python
plt.figure(figsize = (10,5))
plt.boxplot(x= list(df_deaths))
plt.title('Analyzing the total number of deaths happend in Barcelona in the year 2017', fontsize = 30)
plt.xlabel('Deaths in Barcelona', fontsize = 20)
plt.ylabel('Number of Deaths', fontsize = 20)
plt.show()
```

```python
df_district_deaths = df.groupby("District.Name")["Number"].unique()
df_district_deaths
```

```python
plt.figure(figsize = (10,5))
plt.boxplot(x= list(df_district_deaths), labels = list(df_district_deaths.keys()))
plt.xticks(rotation =70)
plt.title('Analyzing the deaths happend in each District', fontsize = 30)
plt.xlabel('Districts', fontsize = 20)
plt.ylabel('Number of Deaths', fontsize = 20)
plt.show()
```

#### We conclude that Eixample has the hight amount of deaths in the year 2017

### Ploting a Histogram

##### - In the first histogram it is shown the density of deaths happend across the country
##### - In the second histogram it is shown the density of deaths happend in each district

```python
plt.figure(figsize = (10,5))
plt.hist(df_deaths, facecolor='red', alpha=0.3, rwidth = 0.95)
plt.title('The Density Plot of deaths happend in Barcelona', fontsize = 20)
plt.xlabel('Range of Deaths')
plt.ylabel('Number of Districts (frequency)')
plt.show()
```

```python
plt.figure(figsize = (10,5))
plt.hist(df_district_deaths, facecolor='red', alpha=0.3, rwidth = 0.9)
plt.title('Deaths happend in each district of Barcelona', fontsize = 20)
plt.xlabel('Range of Deaths')
plt.ylabel('Number of Districts (frequency)')
plt.show()
```
